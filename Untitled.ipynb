```python
import pandas as pd
```

```python
# Lectura del archivo
df = pd.read_csv('glassdoor_jobs.csv') 

# Eliminacion de columna 'Unnamed: 0', la cual es el indice.
df.drop('Unnamed: 0', axis=1, inplace=True) 
```

<b><h3>Salary</h3><b>

```python
# Se crea una variable binaria la cual indica si el salario es por hora(1) o no(0)
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0) 

# Se crea una variable binaria la cual indica si el salario es definido por la empresa(1) o no(0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)
```

```python
# Eliminacion de los valores -1 de la columna 'Salary Estimate'
df = df[df['Salary Estimate'] != '-1']

# Eliminacion del string '(Glassdoor est.)' de la columna 'Salary Estimate'
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])

# Eliminacion de los simbolos 'K' y '$' de la columna 'Salary Estimate'
salary_no_signs = salary.apply(lambda x: x.replace('K','').replace('$',''))

# Eliminacion del string 'per hour' de la columna 'Salary Estimate'
salary_no_perhour = salary_no_signs.apply(lambda x: x.lower().replace('per hour',''))

# Eliminacion del string 'employer provided salary' de la columna 'Salary Estimate'
salary_clean = salary_no_perhour.apply(lambda x: x.lower().replace('employer provided salary:',''))

# La columna 'Salary Estimate' queda con formato min-max
```

```python
# Creacion de la variable 'min_salary' tomando el valor antes del '-'
df['min_salary'] = salary_clean.apply(lambda x: int(x.split('-')[0]))

# Creacion de la variable 'max_salary' tomando el valor despues del '-'
df['max_salary'] = salary_clean.apply(lambda x: int(x.split('-')[1]))

# Creacion de la variable 'avg_salary' con el promedio entre 'min_salary' y 'max_salary'
df['avg_salary'] = (df['min_salary']+df['max_salary'])/2
```

<b><h3>Company name</h3><b>

```python
# Eliminacion del ranking en el nombre de la empresa y el caracter '\n'
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3], axis=1)
df['company_txt'] = df['company_txt'].apply(lambda x: x.replace('\n',''))
```

<b><h3>Location</h3><b>

```python
# Creacion de una variable para hacer split a 'Location' y guardar el estado

def split_job(x):
    response = ''
    try:
        response = x.split(', ')[1]
    except IndexError:
        response = x.split(', ')[0]
    return response

df['job_state'] =  df['Location'].apply(split_job)
df['job_state'] = df['job_state'].replace('Los Angeles', 'CA')
```

```python
# Se crea una variable binaria la cual indica si el trabajo es en los headquarters(1) o no(0)
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
```

<b><h3>Age of company</h3><b>

```python
# Creacion de la variable 'age' que se refiere a la edad de la empresa a partir de la fecha de fundacion
df['age'] = df.Founded.apply(lambda x: x if x <1 else 2020 - x)
```

<b><h3>Jobs description</h3><b>

```python
# Crecion de variables binarias que indican si en 'Job Description' mencionan estas herramientas(1) o no(0)
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
 
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)

df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)

df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)

df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

<b><h3>DF limpio</h3><b>

```python
df_out = df 
df_out.to_csv('salary_data_cleaned.csv',index = False)
```
